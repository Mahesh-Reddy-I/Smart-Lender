# Epic 2: Visualizing and Analyzing the Data

## Story 1: Import and Read Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

import seaborn as sns
```

```python
data=pd.read_csv('../dataset/loan_prediction.csv')
data
```

## Univariate Analysis

Univariate analysis is performed to analyze individual features in the dataset.
Distribution plots are used to understand how numerical variables are distributed.
Here, the distributions of Applicant Income and Credit History are visualized.

```python
#plotting the using distplot
plt.figure(figsize=(12,5))
plt.subplot(121)
sns.distplot(data['ApplicantIncome'],color='r')
plt.subplot(122)
sns.distplot(data['Credit_History'])
plt.show()
```

## Count Plot Analysis

Count plots are used to visualize the frequency of categorical variables in the dataset.
This analysis helps identify the number of occurrences of each category for features such as Gender and Education.

```python
#plotting the count plot
plt.figure(figsize=(18,4))
plt.subplot(1,4,1)
sns.countplot(data['Gender'])
plt.subplot(1,4,2)
sns.countplot(data['Education'])
plt.show()
```

### Observation

- The Gender feature consists of two categories: Male and Female.
- The Education feature consists of Graduate and Not Graduate categories.
- The count plots show the frequency of applicants in each category.
- These visualizations help understand the distribution of categorical variables before data preprocessing.

# Bivariate Analysis

Bivariate analysis is performed to study the relationship between two variables.
Count plots with the **hue** parameter are used to compare categorical features and identify patterns between them.

```python
#visualizing two columns against each other
plt.figure(figsize=(20,5))
plt.subplot(131)
sns.countplot(x=data['Married'],hue=data['Gender'],data=data)
plt.subplot(132)
sns.countplot(x=data['Self_Employed'],hue=data['Education'])
plt.subplot(133)
sns.countplot(x=data['Property_Area'],hue=data['Loan_Amount_Term'])
```

# Multivariate Analysis

Multivariate analysis is used to examine the relationship among multiple variables simultaneously. In this project, a swarm plot is used to visualize the relationship between Applicant Income and Gender while grouping the data based on Loan Status.

```python
#visualized based geder and income what would be the application status
sns.swarmplot(x=data['Gender'],y=data['ApplicantIncome'],hue=data['Loan_Status'],data=data)
```

## Observation

- The swarm plot visualizes the distribution of applicant income across different genders.
- Loan Status (Approved/Rejected) is represented using different colors.
- The plot helps identify how multiple variables interact with one another.
- Multivariate analysis provides deeper insights into the dataset before feature engineering and model training.
