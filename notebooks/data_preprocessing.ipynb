## Checking and Handling Missing Values

```python
# Check the number of missing values in each column
data.isnull().sum()
```

```python
# Fill missing values in Gender with the mode
data['Gender'] = data['Gender'].fillna(data['Gender'].mode()[0])
# Fill missing values in Married with the mode
data['Married'] = data['Married'].fillna(data['Married'].mode()[0])
# Replace '+' in Dependents before filling missing values
data['Dependents'] = data['Dependents'].str.replace('+', '', regex=False)
# Fill missing values in Dependents with the mode
data['Dependents'] = data['Dependents'].fillna(data['Dependents'].mode()[0])
# Fill missing values in Self_Employed with the mode
data['Self_Employed']=data['Self_Employed'].fillna(data['Self_Employed'].mode()[0])
data['LoanAmount'] = data['LoanAmount'].fillna(data['LoanAmount'].mean())
data['Loan_Amount_Term'] = data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mean())
data['Credit_History'] = data['Credit_History'].fillna(data['Credit_History'].mode()[0])
```

```python
# Check whether any missing values remain
data.isnull().sum()
```

## Encoding Categorical Variables

```python
data['Gender']=data['Gender'].map({'Female':1,'Male':0})
data['Property_Area']=data['Property_Area'].map({'Urban':2,'Semiurban':1,'Rural':0})
data['Married']=data['Married'].map({'Yes':1,'No':0})
data['Education']=data['Education'].map({'Graduate':1,'Not Graduate':0})
data['Loan_Status']=data['Loan_Status'].map({'Y':1,'N':0})
data['Dependents'] = data['Dependents'].astype(int)
data['Self_Employed'] = data['Self_Employed'].map({'Yes':1,'No':0})
data.info()
```

## Separating Features and Target Variable

```python
# Remove Loan_ID because it is only an identifier
data = data.drop('Loan_ID', axis=1)
```

```python
# Separate independent variables and target variable
X = data.drop('Loan_Status', axis=1)
y = data['Loan_Status']
```

```python
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

## Splitting the Dataset into Training and Testing Sets

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.33,
    random_state=42,
    stratify=y
)
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

## Scaling the Features

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(
    X_train_scaled,
    columns=X_train.columns,
    index=X_train.index
)

X_test_scaled = pd.DataFrame(
    X_test_scaled,
    columns=X_test.columns,
    index=X_test.index
)
```

## Balancing the Training Dataset Using SMOTE

```python
# Check the class distribution before SMOTE
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)

X_train_balanced, y_train_balanced = smote.fit_resample(
    X_train_scaled,
    y_train
)
```

```python
print("Before SMOTE:")
print(y_train.value_counts())

print("\nAfter SMOTE:")
print(y_train_balanced.value_counts())
```

```python
print("X_train_balanced shape:", X_train_balanced.shape)
print("y_train_balanced shape:", y_train_balanced.shape)
print("X_test shape:", X_test_scaled.shape)
print("y_test shape:", y_test.shape)
```
